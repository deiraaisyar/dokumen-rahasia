```python
import os
import sys
import time
import importlib
from IPython.display import clear_output
from tqdm.auto import tqdm # Progress bar cantik khusus Notebook

# Reload modul untuk memastikan logika terbaru dipakai
import extracting, generating_outputs, save_results
importlib.reload(extracting); importlib.reload(generating_outputs); importlib.reload(save_results)

from extracting import extract_excel, extract_pdf, preprocessing, save_extracted, format_df_to_llm_text
from generating_outputs import process_single_risk, save_cache_to_disk
from save_results import format_and_save_final_excel

clear_output()
print("""
███████╗██╗   ██╗██████╗ ███████╗██████╗     ██╗  ██╗██╗   ██╗██████╗ ██████╗ ██╗██████╗ 
██╔════╝██║   ██║██╔══██╗██╔════╝██╔══██╗    ██║  ██║╚██╗ ██╔╝██╔══██╗██╔══██╗██║██╔══██╗
███████╗██║   ██║██████╔╝█████╗  ██████╔╝    ███████║ ╚████╔╝ ██████╔╝██████╔╝██║██║  ██║
╚════██║██║   ██║██╔═══╝ ██╔══╝  ██╔══██╗    ██╔══██║  ╚██╔╝  ██╔══██╗██╔══██╗██║██║  ██║
███████║╚██████╔╝██║     ███████╗██║  ██║    ██║  ██║   ██║   ██████╔╝██║  ██║██║██████╔╝
╚══════╝ ╚═════╝ ╚═╝     ╚══════╝╚═╝  ╚═╝    ╚═╝  ╚═╝   ╚═╝   ╚═════╝ ╚═╝  ╚═╝╚═╝╚═════╝ 
                 [ BATCH PROCESSING - FULL AUTO-PILOT ]
""")
print("🚀 Sistem AI Paralel akan memproses SELURUH dokumen soal dari folder 'inputs/' ...\n")

start_global = time.time()

# FASE 1: EKSTRAKSI
print("📥 [Tahap 1/5] Mengekstrak data mentah...")
df1, df2, df3, df4 = extract_excel()
df5 = extract_pdf()

print("🧹 [Tahap 2/5] Membersihkan Noise & Preprocessing...")
df1 = preprocessing(df1, "df1"); df2 = preprocessing(df2, "df2")
df3 = preprocessing(df3, "df3"); df4 = preprocessing(df4, "df4"); df5 = preprocessing(df5, "df5")
save_extracted([df1, df2, df3, df4, df5])

# FASE 2: PREDIKSI AI
OUTPUT_DIR = os.path.join(PROJECT_ROOT, "outputs")
dokumen_map = {
    "1. IVC DOE R2": df1, "2. City of York Council": df2, "3. Digital Security IT Sample Register": df3,
    "4. Moorgate Crossrail Register": df4, "5. Corporate Risk Register": df5
}

print("\n🧠 [Tahap 3/5] Memulai Prediksi AI Inference...")

for doc_name, df_clean in dokumen_map.items():
    print("======================================================================")
    print(f"🚀 MEMULAI PROSES FILE: {doc_name}")
    print("======================================================================")
    if df_clean is None or df_clean.empty:
        print(f"⚠️ Skip {doc_name}: Data Kosong.\n")
        continue
        
    llm_texts = format_df_to_llm_text(df_clean)
    final_results = []
    
    # Animasi Progress Bar visual
    for text in tqdm(llm_texts, desc=f"Memprediksi {doc_name[:15]}"):
        res = process_single_risk(target_text=text, project_name=doc_name)
        final_results.append(res)
        
    final_excel_name = f"{doc_name} (Final).xlsx"
    final_excel_path = os.path.join(OUTPUT_DIR, final_excel_name)
    
    try:
        format_and_save_final_excel(final_results, final_excel_path)
        print(f"   ✅ [FINAL OUTPUT] Excel tersimpan di: outputs/{final_excel_name}\n")
    except Exception as e:
        import pandas as pd
        pd.DataFrame(final_results).to_excel(final_excel_path, index=False)
        print(f"   ✅ [FALLBACK OUTPUT] Excel tersimpan di: outputs/{final_excel_name}\n")

try: save_cache_to_disk()
except: pass

print("\n" + "★"*70)
print(f"🎉 SEMUA SOAL LOMBA SELESAI DIPROSES DALAM {time.time() - start_global:.2f} DETIK! 🎉")
print("Silakan cek folder 'outputs/' untuk mengambil 5 File Excel Final!")
print("★"*70)
```
